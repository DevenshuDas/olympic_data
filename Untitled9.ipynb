```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("dataset_olympics.csv")
```

```python
sns.countplot(data=data ,x="Sex",color="y")
plt.title("distribution by gender")
plt.show()
```

```python
sns.histplot(data=data,x="Age",bins=10,kde=True)
plt.title("age distribution")
plt.show()
```

```python
sns.histplot(data=data,x="Height",bins=20,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data=data,x="Weight",bins=20,kde=True,color="g")
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data=data,x="Medal",hue="Medal")
plt.title("medal distribution")
plt.show()
```

```python
sns.countplot(data=data,x="Year",hue="Medal")
plt.title("medal distribution over the year")
plt.xticks(rotation= 90)
plt.show()
```

```python
year_avg_age=data.groupby("Year")["Age"].mean()
print(year_avg_age)
```

```python
sport_median_height=data.groupby("Sport")["Height"].median()
print(sport_median_height.max())
```

```python
country_gender_count=data.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

#most gold country

```python
country_gold=data[data["Medal"]=="Gold"].groupby("NOC")["Medal"].count()
print(country_gold.max())
```

```python
country_gold[country_gold==747]
```

```python
year_participant_count=data.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind="bar")
plt.title("No. of participant over the year")
plt.xlabel("Year")
plt.ylabel("No of participant")
plt.show()
```

```python
country_avg_age=data.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.head(10).plot(kind="bar",color="g")
plt.title("top 10 countries with highest average age of participants")
plt.xlabel("country")
plt.ylabel("avg_age")
plt.show()
```

```python
country_avg_age=data.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.tail(10).plot(kind="bar")
plt.title("top 10 countries with lowest average age of participants")
plt.xlabel("country")
plt.ylabel("avg_age")
plt.show()
```

```python
sns.boxplot(data=data,x="Season",y="Age",color="r")
plt.title("Distribution of Age by Seasona")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(data=data,x="Medal",y="Height",color="b")
plt.title("distribution of medal by height")
plt.xlabel("medal")
plt.ylabel("height")
plt.show()
```

```python
most_medal_country=data["NOC"].value_counts().idxmax()
print("Most_medal_country:",most_medal_country)
```

```python
tallest_athlete=data[data["Height"]==data["Height"].max()]
print("Tallest Athlete:")
print(tallest_athlete[["ID","Name","Height","Sport"]])
```

```python
heaviest_athlete=data[data["Weight"]==data["Weight"].max()]
print("Heaviest athlete:")
print(heaviest_athlete[["ID","Name","Height","Sport"]])
```

```python
sns.scatterplot(data=data,x="Height",y="Weight",hue="Medal")
plt.title("Athlete height vs weight by medal status")
plt.xlabel("height")
plt.ylabel("weight")
plt.legend(title="Medal")
plt.show()
```
